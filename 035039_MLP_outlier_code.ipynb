```python
import pandas as pd
import numpy as np

# Iterating through all columns in the dataframe
for column in df.columns:
    # Check if the column is not already an outlier column
    if '_outlier' not in column:
        # Calculate the 1st and 3rd quartile
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        # Interquartil range
        IQR = Q3 - Q1
        # Creates a new column with the name of the column + "_outlier"
        df[column + "_outlier"] = np.where((df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR)), -1, 0)
        df[column + "_outlier"] = np.where((df[column] > (Q3 + 1.5 * IQR)), 1, df[column + "_outlier"])
        
print(df)
```
