__ABOUT THE DATA SET__

1. The data has been picked up from the *Amazon seller account of KHOJI* (a company selling virgin plastic planters). 


2. The data has 91*23 rows and coloumns, respectively and describes the page sessions, page views, orders and the total sales of each 93 unique SKUs of the company over a period of 1 year (Jan 2022- Dec-2022).

4. __Quantitative varibales__: Following are the quantitaive varibales which have been further bifurcated into B2B and normal customers 

     a. *Sessions*: The number of unique visitors to your product page

     b. *Page Views*: Total number of times the product page was looked at (this can include multiple/repeat visits from the same people)

     c. *Units orders*: Total number of units sold

     d. *Ordered product sales*: Total number of units sold

     e. *Total Order Items*: Total orders that contained at least one of this SKU

     f. *Buy Box Percentage*: When a visitor landed on the product page, how often were the seller featured in the buy box (ADD TO CART)

     g. *Unit Session Percentage*: Ratio of units ordered to sessions (very similar to a conversion rate)

```python
#Get overall total for units ordered and product sales 
def column_total(df,columns):
    return df[columns].sum(axis=0)

df1_columns = ['Units Ordered','Ordered Product Sales' ]
df1_total = column_total(df1, df1_columns)
df2_columns = ['Units Ordered - B2B','Ordered Product Sales - B2B']
df2_total = column_total(df2, df2_columns)

print('Normal Overall:',df1_total)
print('B2B Overall:',df2_total)
```

```python
#Get total for units ordered and product sales for category
def get_ctg_total(df, columns, Class='Hanging Pots'):
    return df[df['Class']==Class][columns].sum()

Class = 'Hanging Pots'
df1_columns = ['Units Ordered', 'Ordered Product Sales']
df2_columns = ['Units Ordered - B2B', 'Ordered Product Sales - B2B']
ctg='Hanging Pots'
norm_total = get_ctg_total(df1, df1_columns, Class)
b2b_total = get_ctg_total(df2, df2_columns, Class)
print(f'{ctg} Total:', norm_total)
print(f'{ctg} Total (B2B):', b2b_total)
```

__GENERATE CARDS FOR OVERALL NUMBERS__

```python
def generate_card_content(card_header, overall_value):
    card_head_style = {'textAlign':'center','fontSize':'150%'}
    card_body_style = {'textAlign':'center','fontSize':'200%'}
    card_header = dbc.CardHeader(card_header,style=card_head_style)
    card_body = dbc.CardBody(
        [
            html.H5(f"{int(overall_value):,}", className="card-title",style=card_body_style),
            html.P(
                #"Overall: {:,}".format(overall_value[0].iloc[0]),
                className="card-text",style={'textAlign':'center'}
            ),
        ]
    )
    card = [card_header,card_body]
    return card
```

```python
def generate_cards():
    df = pd.read_csv(r"C:\Users\Prabhleen Kaur Pahwa\Documents\1_FORE\Term 3\MLP\KHOJI DATA_FINAL.csv", encoding='cp1252')
    df_column1 = ['Units Ordered']
    df_column2 = ['Ordered Product Sales']
    df_column3 = ['Units Ordered - B2B']
    df_column4 = ['Ordered Product Sales - B2B']
    df_total1 = column_total(df, df_column1).sum()
    df_total1 = df_total1.astype(float)
    df_total2 = column_total(df, df_column2).sum()
    df_total2 = df_total2.astype(float)
    df_total3 = column_total(df, df_column3).sum()
    df_total3 = df_total3.astype(float)
    df_total4 = column_total(df, df_column4).sum()
    df_total4 = df_total4.astype(float)
    cards = html.Div(
        [
            dbc.Row(
                [
                    dbc.Col(dbc.Card(generate_card_content("Total Sales", df_total2), color="success", inverse=True),md=dict(size=2,offset=3)),
                    dbc.Col(dbc.Card(generate_card_content("Total Orders", df_total1), color="warning", inverse=True),md=dict(size=2)),
                    dbc.Col(dbc.Card(generate_card_content("Total Sales B2B", df_total4), color="success", inverse=True),md=dict(size=2)),
                    dbc.Col(dbc.Card(generate_card_content("Total Orders B2B", df_total3), color="warning", inverse=True),md=dict(size=2))
                ],
                className="mb-4",
            ),
        ],id='card1'
    )
    return cards
```
