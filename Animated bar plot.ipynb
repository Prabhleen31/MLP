__ABOUT THE DATA SET__

1. The data has been picked up from the *Amazon seller account of KHOJI* (a company selling virgin plastic planters). 


2. The data has 91*23 rows and coloumns, respectively and describes the page sessions, page views, orders and the total sales of each 93 unique SKUs of the company over a period of 1 year (Jan 2022- Dec-2022). 


3. __Descriptive variables__: The rows contain 93 unique SKU's of KHOJI, along with 93 unique Parent ASINs and titles. Further, the products are broadly divided into 5 classes:

      a. Hanging Pots

      b. Planters

      c. Decorative (indoor) Planters

      d. Trays

      e. Water cans 



4. __Quantitative varibales__: Following are the quantitaive varibales which have been further bifurcated into B2B and normal customers 

     a. *Sessions*: The number of unique visitors to your product page

     b. *Page Views*: Total number of times the product page was looked at (this can include multiple/repeat visits from the same people)

     c. *Units orders*: Total number of units sold

     d. *Ordered product sales*: Total number of units sold

     e. *Total Order Items*: Total orders that contained at least one of this SKU

     f. *Buy Box Percentage*: When a visitor landed on the product page, how often were the seller featured in the buy box (ADD TO CART)

     g. *Unit Session Percentage*: Ratio of units ordered to sessions (very similar to a conversion rate)
